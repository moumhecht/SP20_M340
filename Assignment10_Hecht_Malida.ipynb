**(Problem 1)** Find the Lagrangian Interpolating polynomial of degree three that passes through the points in the given table:

The Lagrange Interpolating Polynomial is given by:
$$f_n (x)= \sum _{i=0}^n L_i(x) f(x_i)$$
Where $L_i(x)$ is defined as:
$$L_i(x) = \prod_{j = 0,j\neq i}^n \frac{x-x_j}{x_i-x_j}$$

We have 4 data points, so $n=3$. Then our lagrange polynomial can be expanded to be:
$$\begin{aligned}
f_3(x) &= L_0(x)f(x_0) + L_1(x) f(x_1) + L_2(x) f(x_2)+ L_3(x) f(x_3) \\
& = 227.04 L_0(x) + 362.78 L_1(x) + 517.35 L_2(x) + 602.97 L_3(x)
\end{aligned}$$
Then, the weighting factors can be written as:
$$\begin{aligned}
L_0(x) & = \bigg(\frac{x - 362.78}{227.04-362.78}\bigg)\bigg(\frac{x - 517.35}{227.04-517.35}\bigg)\bigg(\frac{x -602.97}{227.04-602.97}\bigg) \\
&=  \bigg(\frac{x - 362.78}{-135.74}\bigg)\bigg(\frac{x - 517.35}{-290.31}\bigg)\bigg(\frac{x -602.97}{-375.93}\bigg) \\
&= - \bigg(\frac{x - 362.78}{135.74}\bigg)\bigg(\frac{x - 517.35}{290.31}\bigg)\bigg(\frac{x -602.97}{375.93}\bigg) \\
L_1(x) & = \bigg(\frac{x - 227.04}{362.78-227.04}\bigg)\bigg(\frac{x - 517.35}{362.78-517.35}\bigg)\bigg(\frac{x -602.97}{362.78-602.97}\bigg) \\
&=  \bigg(\frac{x -227.04}{135.74}\bigg)\bigg(\frac{x - 517.35}{-154.57}\bigg)\bigg(\frac{x -602.97}{-240.19}\bigg) \\
&=  \bigg(\frac{x -227.04}{135.74}\bigg)\bigg(\frac{x - 517.35}{154.57}\bigg)\bigg(\frac{x -602.97}{240.19}\bigg) \\
L_2(x) & = \bigg(\frac{x - 227.04}{517.35-227.04}\bigg)\bigg(\frac{x - 362.78}{517.35-362.78}\bigg)\bigg(\frac{x -602.97}{517.35-602.97}\bigg) \\
&=  \bigg(\frac{x - 227.04}{290.31}\bigg)\bigg(\frac{x - 362.78}{154.57}\bigg)\bigg(\frac{x -602.97}{-85.62}\bigg) \\
&= - \bigg(\frac{x - 227.04}{290.31}\bigg)\bigg(\frac{x - 362.78}{154.57}\bigg)\bigg(\frac{x -602.97}{85.62}\bigg) \\
L_3(x) & = \bigg(\frac{x - 227.04}{602.97-227.04}\bigg)\bigg(\frac{x - 362.78}{602.97-362.78}\bigg)\bigg(\frac{x -517.35}{517.35-517.35}\bigg) \\
&=  \bigg(\frac{x - 227.04}{375.93}\bigg)\bigg(\frac{x - 362.78}{240.19}\bigg)\bigg(\frac{x -517.35}{85.62}\bigg) \\
&=  \bigg(\frac{x -227.04}{375.93}\bigg)\bigg(\frac{x - 362.78}{240.19}\bigg)\bigg(\frac{x -517.35}{85.62}\bigg) \\
\end{aligned}$$
Then the third degree Lagrange Interpolating polynomial is:
$$\begin{aligned}
f_3(x) &= L_0(x)f(x_0) + L_1(x) f(x_1) + L_2(x) f(x_2)+ L_3(x) f(x_3) \\
& = 227.04 L_0(x) + 362.78 L_1(x) + 517.35 L_2(x) + 602.97 L_3(x)\\
& =  -227.04 \bigg(\frac{x - 362.78}{135.74}\bigg)\bigg(\frac{x - 517.35}{290.31}\bigg)\bigg(\frac{x -602.97}{375.93}\bigg)
+ 362.78\bigg(\frac{x -227.04}{135.74}\bigg)\bigg(\frac{x - 517.35}{154.57}\bigg)\bigg(\frac{x -602.97}{240.19}\bigg) \\&
-517.35\bigg(\frac{x - 227.04}{290.31}\bigg)\bigg(\frac{x - 362.78}{154.57}\bigg)\bigg(\frac{x -602.97}{85.62}\bigg) + 602.97 \bigg(\frac{x -227.04}{375.93}\bigg)\bigg(\frac{x - 362.78}{240.19}\bigg)\bigg(\frac{x -517.35}{85.62}\bigg) 
\end{aligned}$$

**(Problem 2.ii)** Use the python functions to compute $f(x)$ for the $x$ generated in problem (2.i)

```python
def L_weight(x_i,j,x):
    L_i = np.ones(x.size)
    n = x_i.size
    for i in range(0,n):
        if i !=j:
            L_i *= ((x-x_i[i]) / float(x_i[j]-x_i[i]))
    return L_i
```

```python
def Lagrange(x_i,f_i,x):
    n = f_i.size
    polynomial_i = np.zeros(x.size)
    for k in range(0,n):
        polynomial_i +=f_i[k]*L_weight(x_i,k,x)
    return polynomial_i
```

```python
exes = np.array([10,15,20,22.5])
effs = np.array([227.04,362.78,517.34,602.94])
eff_interpolated = Lagrange(exes,effs,x)
plt.plot(x,eff_interpolated,'*')
plt.plot(exes,effs,'*')
plt.title('Third Degree Lagrange Polynomial')
plt.xlabel('$x$')
plt.ylabel('$P_3(x)$')
```

**(Problem 5)** An automobile of mass $M = 5400$ kg is moving at a speed of $30$ m/s. The engine is disengaged suddenly at $t=0$ sec. Assume that the equation of motion after t=0 is given by

$$
5400 v \frac{dv}{dx} = -8.276v^2 -2000
$$

where $v=v(t)$ is the speed (m/sec) of the car at $t$. The left side represents $Mv(dv/dx)$. The first term on the right is the aerodynamic drag, and the second term is the rolling resistance of the tires. Calculate how far the car travels until the speed reduces to 15 m/sec.

Hint: The equation of motion may be integrated as:

$$
\int_{15}^{30} \frac{5400 v}{8.276v^2 +2000}\ dv = \int dx = x
$$

Evaluate the equation above using Simpson's rule.

```python
def simpson(a,b,n,f):
    n_int = int(2*n) # This makes N even incase some one enters an odd value for N
    exes = np.linspace(a,b,n_int+1)
    effs = f(exes)
    dx = (b-a)/n_int
    return dx/3.*(effs[0] + effs[n_int] + 2.*np.sum(effs[2:n_int-1:2]) + 4.*np.sum(effs[1:n_int:2])) 
```

```python
func = lambda v: (5400*v)/( (8.276*v**2)+2000 )
```

```python
print('The car travels %1.5f m before reaching v = 15 m/sec' %simpson(15,30,1e3,func))
```

**(Problem 6)** The Hermite Polynomials $H_n(x)$ are such that $H_0(x) = 1$, $H_1(x) = 2x$, and for $n\geq2$, the recurrence relationship
$$H_n(x) = 2x H_{n-1}(x) - 2(n-1)H_{n-2}(x) $$

**(Problem 6.a)** Write a recursive program to find any Hermite polynomial $H_n(x)$

```python
def hermite(x,n):
    H0 = 1
    H1 = 2*x

    if n == 0:
        return H0
    elif n == 1:
        return H1
    else:
        Hn1 = H1
        Hn2 = H0
        for k in range (2,n+1):
            Hn = (2*x*Hn1) - (2*(n-1)*Hn2)
            Hn2 = Hn1
            Hn1 = Hn
        return Hn
```

**(Problem 6.b)** Generate a plot of the first three Hermite polynomials on the interval $-2\leq x \leq 2$. Your figure should be labeled, have a legend, and different colors should be used to distinguish the different Hermite polynmials in the plot.

```python
exes = np.linspace(-2,2,int(1e3))

whys0 = [hermite(i,0) for i in exes]
plt.plot(exes,whys0, label = 'n=0', c = 'm')

whys1 =  [hermite(i,1) for i in exes]
plt.plot(exes,whys1, label = 'n=1', c = 'b')

whys2 =  [hermite(i,2) for i in exes]
plt.plot(exes,whys2, label = 'n=2', c = 'k')


plt.xlabel('x')
plt.ylabel('$H_n(x)$')
plt.legend(loc=(.7,.75))
plt.title('First three Hermite Polynomials')
```
